```python
import sqlite3
```

```python
conn = sqlite3.connect("chinook.db")
```

```python
cursor = conn.cursor()
```

```python
query_2_1a = """
SELECT InvoiceId,
r.FirstName || ' ' || r.LastName AS 'Nombre',
BillingCountry
FROM invoices l
INNER JOIN customers r ON
r.CustomerId == l.CustomerId
WHERE l.BillingCountry == 'Brazil'
ORDER BY InvoiceId;
"""
cursor.execute(query_2_1a)
cursor.fetchall()
```

```python
query_2_1b = """
SELECT InvoiceId,
r.FirstName || ' ' || r.LastName AS 'Nombre',
BillingCountry,
r.Country
FROM invoices l
INNER JOIN customers r ON
r.CustomerId == l.CustomerId
WHERE r.Country == 'Brazil'
ORDER BY InvoiceId;
"""
cursor.execute(query_2_1b)
cursor.fetchall()
```

```python
query_check_country = """
SELECT *
FROM customers c, invoices i
WHERE c.customerId = i.customerId AND c.country <> i.BillingCountry;
"""
cursor.execute(query_check_country)
cursor.fetchall()
```

```python
query_3 = """
SELECT e.FirstName || ' ' || e.Lastname as 'Nombre completo', i.* 
FROM invoices i
JOIN customers c ON i.CustomerId == c.CustomerId
JoIN employees e ON c.SupportRepID == e.EmployeeId
"""
cursor.execute(query_3)
cursor.fetchall()
```

```python
# 4. Muestra el nombre del cliente, el pais, el nombre del agente y el total 
query_4 = """
SELECT c.FirstName || ' ' || c.Lastname as 'Nombre completo cliente', c.Country,  e.FirstName || ' ' || e.Lastname as 'Nombre completo agente', i.Total 
FROM invoices i
JOIN customers c ON i.CustomerId == c.CustomerId
JoIN employees e ON c.SupportRepID == e.EmployeeId;
"""

cursor.execute(query_4)
cursor.fetchall()
```

```python
# 5. Muestra cada articulo de la factura con el nombre de la cancion 
query_5 = """
SELECT it.*, tr.Name
FROM invoice_items it
JOIN tracks tr ON it.TrackId == tr.TrackId
"""
cursor.execute(query_5)
cursor.fetchall()
```

```python
# 6. Muestra todas las canciones con su nombre, formato, album y genero  
query_6 = """
SELECT tr.Name AS Track, m.Name as Media, a.Title AS Album, g.Name as Genre
FROM tracks tr
JOIN media_types m ON tr.MediaTypeId == m.MediaTypeId
JOIN albums a ON tr.AlbumId == a.AlbumId
JOIN genres g ON tr.GenreId == g.GenreId
"""
cursor.execute(query_6)
cursor.fetchall()
```

```python
# 7. Muestra cuantas canciones hay en cada playlist y el nombre de cada playlist  
query_7 = """
SELECT  p.Name as Playlist, COUNT(ptr.PlaylistId) AS Tracks
FROM playlists p
LEFT JOIN playlist_track ptr  ON ptr.PlaylistId == p.PlaylistId
GROUP BY p.PlaylistId
ORDER BY Tracks DESC;
"""
cursor.execute(query_7)
cursor.fetchall()
```

```python
# 8. Muestra cuánto ha vendido cada empleado   
query_8 = """
SELECT e.FirstName || ' ' || e.LastName AS Empleado, ROUND(SUM(i.Total), 2) AS Total
  FROM invoices i
  JOIN customers c ON i.CustomerId == c.CustomerId
  JOIN employees e ON c.SupportRepId == e.EmployeeId
  GROUP BY e.EmployeeId;
"""
cursor.execute(query_8)
cursor.fetchall()
```

```python
# 9. Quien ha sido el agente de ventas que más ha vendido en 2009? 
query_9 = """
SELECT e.FirstName || ' ' || e.LastName AS Empleado, SUM(i.Total) AS Total
FROM invoices i
JOIN customers c ON i.CustomerId == c.CustomerId
JOIN employees e ON c.SupportRepId == e.EmployeeId
WHERE i.InvoiceDate LIKE '2009%'
GROUP BY e.EmployeeId
ORDER BY SUM(i.Total) DESC
LIMIT 1;

"""

cursor.execute(query_9)
cursor.fetchall()
```

```python
# 10. Quienes son los 3 grupos que más han vendido?
query_10 = """
SELECT at.Name, ROUND(SUM(it.UnitPrice * it.Quantity),2) AS Total
FROM invoice_items it
JOIN tracks tr ON it.TrackId == tr.TrackId
JOIN albums a ON tr.AlbumId == a.AlbumId
JOIN artists at ON a.ArtistId == at.ArtistId
GROUP BY at.Name 
ORDER BY Total DESC
LIMIT 3;
"""

cursor.execute(query_10)
cursor.fetchall()
```
