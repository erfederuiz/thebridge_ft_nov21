9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.

```python
x = ['Ana', 'Luis', 'Pedro', 'Sofia', 'Carmen']
y = [160, 180, 175, 190, 170]

plt.style.use('seaborn-white')
barplot = plt.bar(x,y,align='center', color='g') 
plt.xlabel('Amigos')
plt.ylabel('Alturas')
for bar in barplot:
    plt.text(bar.get_x()+0.30,bar.get_height()+7,str(bar.get_height()), va='top')


plt.show()
```

11. Bas치ndote en el gr치fico anterior, borra la escala del eje Y (ya aparece en la altura)

```python
x = ['Ana', 'Luis', 'Pedro', 'Sofia', 'Carmen']
y = [160, 180, 175, 190, 170]

plt.style.use('seaborn-white')
plt.figure(figsize=(3,5), frameon = False)
plt.xlabel('Amigos')
plt.ylabel('Alturas')
plt.yticks([])
plt.xticks(rotation=90)
plt.box(False)
barplot = plt.bar(x,y,align='center', color='g') 
for bar in barplot:
    plt.text(bar.get_x(),bar.get_height()+7,str(bar.get_height()), va='top')


plt.show()
```
