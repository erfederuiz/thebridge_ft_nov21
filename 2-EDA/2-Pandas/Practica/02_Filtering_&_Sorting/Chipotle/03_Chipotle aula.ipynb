### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
address = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

### Step 3. Assign it to a variable called chipo.

```python
chipo  = pd.read_csv(address, sep='\t', low_memory=False)
```

```python
chipo.head(25)
```

```python
chipo.shape
```

```python
chipo['money_replace'] = chipo['item_price'].str.replace('$', '')
chipo['money_replace'] = pd.to_numeric(chipo['money_replace'])
```

### Step 4. How many products cost more than $10.00?

```python
more_10 = chipo["money_replace"].astype(dtype ='float64') > 10
#Puede haber varios pedidos de un mismo producto
chipo[more_10]["item_name"].nunique()
```

### Step 6. Sort by the name of the item

### Step 7. What was the quantity of the most expensive item ordered? 2 ways

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo.loc[chipo['item_name'] == 'Veggie Salad Bowl',['item_name']].count()[0]
```

```python
chipo.groupby("item_name").sum().loc["Veggie Salad Bowl"]
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_mask = chipo['item_name'] == 'Canned Soda' 
canned_quantity = chipo['quantity'] > 1
chipo.loc[canned_mask & canned_quantity,['item_name']].count()[0]
```

```python
len(chipo.loc[canned_mask & canned_quantity,['item_name']])
```
