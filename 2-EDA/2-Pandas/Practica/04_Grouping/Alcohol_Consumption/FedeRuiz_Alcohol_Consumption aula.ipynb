### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a variable called drinks.

```python
address = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(address, low_memory=False)
drinks.describe()
```

```python
drinks.dropna(axis=0, inplace=True)
```

### Step 4. Which continent drinks more beer on average?

```python
s = drinks.groupby('continent')['beer_servings'].mean()
s
```

```python
s = drinks.groupby('continent').mean()
s[s['beer_servings']]
```

```python
continents = drinks.groupby('continent')
```

```python
continents.groups
```

```python
continents.get_group('EU').describe()
```

```python
continents.sum()
```

```python
sum_continents = continents.sum()
sum_column = sum_continents['beer_servings'] + sum_continents['spirit_servings'] + sum_continents['wine_servings'] 
sum_continents['total_servings'] = sum_column
beer_over_total = sum_continents['beer_servings'] / sum_continents['total_servings']
wine_over_total = sum_continents['wine_servings'] / sum_continents['total_servings']
spirit_over_total = sum_continents['spirit_servings'] / sum_continents['total_servings']
sum_continents['beer_over_total'] = beer_over_total
sum_continents['wine_over_total'] = wine_over_total
sum_continents['spirit_over_total'] = spirit_over_total
sum_continents[sum_continents['beer_over_total'] == sum_continents['beer_over_total'].max()]
```
